```python
import os
from dotenv import load_dotenv

# Add Azure OpenAI package
from openai import AzureOpenAI

def openai_test_3(): 
        
    try: 
    
        # Get configuration settings 
        load_dotenv()
        azure_oai_endpoint = os.getenv("AZURE_OAI_ENDPOINT")
        azure_oai_key = os.getenv("AZURE_OAI_KEY")
        azure_oai_deployment = os.getenv("AZURE_OAI_DEPLOYMENT")
        
        # Initialize the Azure OpenAI client...
        client = AzureOpenAI(
        azure_endpoint=azure_oai_endpoint,
        api_key=azure_oai_key,
        api_version="2025-01-01-preview",
        )

        messages = [
            {
                "role" : "system",
                "content" : [
                {
                    "type" : "text",
                    "text" : "너는 사용자가 입력한 값에 대해서 설명해주고, 답변은 너무 길지않게 해"
                }
                ]
            }
        ]

        while True:
            # Get input text
            input_text = input("질문을 입력하세요 ('quit' 이라고 입력하면 종료): ")
            if input_text.lower() == "quit":
                break
            if len(input_text) == 0:
                print("질문을 다시 작성해주세요.")
                continue

            print("질문 : " + input_text)

            # print("\n답변하는중...\n\n"
            
            messages.append(
                {
                    "role" : "user",
                    "content" : [
                    {
                        "type" : "text",
                        "text" : input_text

                    }
                    ]
                }
            )

            # 나는 한 입력-출력 당 user 와 assistant를 둘 다 추가하니 이전 3개까지 기억하기 위해 system값 포함 8개를 기억함
            # 시스템(1개) + 현재 나의 입력(1개) + 이전 3개의 입력-출력 쌍(6개) = 8개
            if len(messages) > 8 :
                messages = [messages[0]] + messages[-7:]
            
            # Add code to send request...
            completion = client.chat.completions.create(
            model=azure_oai_deployment,
            messages=messages,
            max_tokens=30,
            temperature=0.7,
            top_p=0.95,
            frequency_penalty=0,
            presence_penalty=0,
            stop=None,
            stream=False
            )
            
            print("답변 : " + completion.choices[0].message.content + "\n")

            messages.append(
                {
                    "role" : "assistant",
                    "content" : [
                    {
                        "type" : "text",
                        "text" : completion.choices[0].message.content

                    }
                    ]
                }
            )
            

    except Exception as ex:
        print(ex)
```

```python
openai_test_3()
```
